cvxlab - ROLLING HORIZON TEST

```python

from data_handler_multistage import *
import cvxlab as cl

H_group = {
    1: range(2000, 2030),
    2: range(2005, 2030),
    3: range(2010, 2030),
    4: range(2015, 2030),
    5: range(2020, 2030),
    6: range(2025, 2030),
}

main_dir_path = 'C:/Users/baioc/OneDrive - Politecnico di Milano/PhD/CVXLAB/Concepts/Rolling - v2'
fh_len = 5
delta = 6



```

RH1: 2000 - 2030

```python
H=1

model_dir_name = f'H{H}'


pass_model_settings(H, main_dir_path, start = True)
update_sets(H, H_group[H], main_dir_path, delta,start=True)
update_input_data(H, H_group[H], main_dir_path, delta, fh_len, start=True)

#define model instance
model1 = cl.Model(
    model_dir_name=model_dir_name,
    main_dir_path=main_dir_path,
    log_level='debug',
    model_settings_from='xlsx',
    multiple_input_files=True,
    use_existing_data=False,
    detailed_validation=True,
)
#load data
model1.load_model_coordinates()
model1.initialize_blank_data_structure()
model1.load_exogenous_data_to_sqlite_database(force_overwrite=True)
#initialize model and run
# model.update_database_and_problem()
model1.initialize_problems()
model1.run_model(
    verbose=False,
    solver='GUROBI',
    integrated_problems=False,
)

#save results
model1.load_results_to_database(force_overwrite=True)
```
